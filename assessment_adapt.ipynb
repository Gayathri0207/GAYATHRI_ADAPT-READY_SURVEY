2.Given an unsorted array of integers, find the length of the longest continuous increasing subsequence (subarray). 

```python

def find_length(nums):
    if not nums:
        return 0

    max_length = 1
    current_length = 1

    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_length += 1
            max_length = max(max_length, current_length)
        else:
            current_length = 1

    return max_length

input1 = [1, 3, 5, 4, 7]
output = find_length(input1)
print("Output for Input 1:", output)



```

3. Given a list of non negative integers, arrange them such that they form the largest number.

```python
from typing import List

def largestNumber(nums: List[int]) -> str:
    # Custom comparison function for sorting
    def compare(x, y):
        return int(y + x) - int(x + y)

    # Convert numbers to strings and sort using custom comparison
    nums = sorted(map(str, nums), key=lambda x: (x[0], x), reverse=True)

    # Concatenate the sorted numbers
    largest_num = ''.join(nums)

    # Remove leading zeros if any
    largest_num = largest_num.lstrip('0')

    # If the number becomes empty, it means all numbers were zeros
    return largest_num if largest_num else '0'

# Example usage:
input1 = [10, 2]
output = largestNumber(input1)
print("Output for Input 1:", output)

```

4. Store all the "servlet-name", and "servlet-class" to a csv file from the attached sample_json.json file using Python.

```python
import json
import csv

# Reading JSON file
with open('sample_json.json', 'r') as json_file:
    data = json.load(json_file)

# Extracting "servlet-name" and "servlet-class" from JSON
servlets = [(servlet['servlet-name'], servlet['servlet-class']) for servlet in data['web-app']['servlet']]

# CSV file
with open('servlet_data.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    # Write header
    writer.writerow(['servlet-name', 'servlet-class'])
    # Write rows
    writer.writerows(servlets)

print("CSV file generated")
```
